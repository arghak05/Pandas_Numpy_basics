## Pandas

```python
import pandas as pd
import numpy as np 
```

```python
np.random.seed(20)
df = pd.DataFrame(np.random.randn(1000,1) , index=pd.date_range('2021-07-18' , periods=1000), columns=['data'])
```

```python
df
```

```python
df.data.cumsum()
```

```python
df['data1'] = df.data.cumsum()
```

```python
df
```

```python
df['data_power_3'] = df['data'].apply(lambda x : x ** 3)
```

```python
df
```

```python
def test(x):
    if x < 0:
        return x
    else:
        return np.log(x)
```

```python
df['data_log'] = df['data'].apply(test)
```

```python
df
```

```python
def test1(x):
    if x < 0:
        return 1
    else:
        return 0
```

```python
df['data_flag'] = df['data'].apply(test1)
```

```python
df
```
